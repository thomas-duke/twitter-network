```python
import networkx as nx
from networkx.algorithms.community import *
import matplotlib
import seaborn as sns
```

```python
G = nx.les_miserables_graph()
```

```python
def draw_communities(communities):
    communities = list(communities)
    if is_partition(G, communities):
        print('Coverage: {}'.format(coverage(G, communities)))
        print('Performance: {}'.format(performance(G, communities)))
    pos=nx.kamada_kawai_layout(G) 
    colors = sns.color_palette()
    for i in range(len(communities)):
        nx.draw_networkx_nodes(G, pos, nodelist = list(communities[i]), node_color = [colors[i]])
    nx.draw_networkx_edges(G,pos,width=1.0,alpha=0.5)
```

```python
draw_communities(greedy_modularity_communities(G))
```

```python
draw_communities(asyn_fluidc(G, k = 3))
```

```python
gn = girvan_newman(G)
draw_communities(list(set(c) for c in next(gn)))
```
